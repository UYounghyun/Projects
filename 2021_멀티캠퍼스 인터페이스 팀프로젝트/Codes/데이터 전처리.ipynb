```python
import pymysql
import numpy as np
import pandas as pd
```

```python
street_data = pd.read_csv('data/data/street.csv', encoding='CP949')
for i in range(len(street_data)):
    street_data.loc[i:,['상세주소']]= '{"상세주소":"'+street_data['지번 주소'][i]+'", "x":'+str(street_data['중심 좌표 X'][i])+', "y":'+str(street_data['중심 좌표 Y'][i])+'}'
```

```python
street_df = street_data[['최종 표기명', '행정 구', '상세주소']].copy()
street_df.columns = ['명칭', '행정 구', '상세주소']
street_df['분류3'] = '거리'
display(street_df)
```

```python
museum_data = pd.read_csv('data/data/museum.csv', encoding='CP949')
museum_df = museum_data[['명칭','행정구','분류3']].copy()
museum_df.columns = ['명칭', '행정 구','분류3']
display(museum_df)
```

```python
nature_data = pd.read_csv('data/data/nature.csv', encoding='CP949')
nature_df = nature_data[['명칭', '행정 구']].copy()
nature_df['분류3'] = '공원'
display(nature_df)
```

```python
park_data = pd.read_csv('data/data/park.csv', encoding='CP949')
park_df = park_data[['명칭', '행정 구','분류3']].copy()
park_df.loc[(park_df['분류3'] == '휴양림') | (park_df['분류3'] == '식물원'),'분류3'] = '수목원'
display(park_df)
```

```python
history_data = pd.read_csv('data/data/history.csv', encoding='CP949')
history_df = history_data[['명칭','행정 구','분류3']].copy()
history_df['분류3'] = '유적지'
display(history_df)
```

```python
# 구 데이터
gu_data = ['종로구','중구','용산구','성동구','광진구','동대문구','중랑구','성북구','강북구','도봉구','노원구','은평구','서대문구','마포구','양천구','강서구','구로구','금천구','영등포구','동작구','관악구','서초구','강남구','송파구','강동구']
gu_df = pd.DataFrame([gu_data,np.arange(1,26)])
gu_df=gu_df.T
gu_df.columns=['행정 구', '행정구 번호']
display(gu_df)
```

```python
# 카테고리 데이터
category_data = ['수목원', '동물원', '유적지', '박물관', '거리', '공원']
category_df = pd.DataFrame([category_data,np.arange(1,7)])
category_df = category_df.T
category_df.columns = ['분류3','카테고리 번호']
display(category_df)
```

```python
# 관광지 데이터 결합
raw_data = pd.concat([museum_df, nature_df, park_df, history_df])
display(raw_data)
# 관광데이터와 행정구, 카테고리 데이터 머지
# 머지하면서 서울의 행정구에 없는 관광지 몇개 발견되어서 제외함 (8개소)
data_gu = pd.merge(raw_data, gu_df, on='행정 구')
data = pd.merge(data_gu, category_df, on='분류3')

street_data = pd.merge(street_df, gu_df, on='행정 구')
street_df = pd.merge(street_data, category_df, on='분류3')
display(data)
display(street_df)
```

```python
# 테스트 코드
keyword = '관악디자인거리'
url = 'https://dapi.kakao.com/v2/local/search/keyword.json?&query=' + keyword
result = requests.get(urlparse(url).geturl(), headers={'Authorization': 'KakaoAK c1c91e71e21bbee8cdc83ab7f30cbc51'}).json()
print(result)
```

```python
for a in data['명칭']:
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?&query=' + a
    result = requests.get(urlparse(url).geturl(), headers={'Authorization': 'KakaoAK c1c91e71e21bbee8cdc83ab7f30cbc51'}).json()
    try:
        data.loc[data['명칭']==a, ['상세주소']] = result['documents'][0]['address_name']
    except IndexError :
        data.loc[data['명칭']==a, ['상세주소']] = np.nan
```

```python
data= pd.concat([data, street_df])
# 주소 없는 것 추려냄
data = data[data.isnull()['상세주소']==False]
display(data)
```

- 상세정보 추가(wikipedia)

```python
import wikipedia as wk
wk.set_lang("ko")
```

```python
# 테스트 코드
wk.summary('홍릉수목원')
```

```python
n=0
for a in data['명칭']:
    try:
        data.loc[data['명칭']== a, ['소개']] = wk.summary(a)
    
    except wk.exceptions.DisambiguationError as dis :
        pass
    
    except wk.exceptions.PageError :
        data.loc[data['명칭']== a, ['소개']] = a
    
    finally :
        print(n)
        n+=1
```

```python
# 검색 안되는거 제외
raw_data = data[data.isnull()['소개']==False]
display(raw_data)
```

```python
# 최종 결과 저장
raw_data.reset_index().to_csv('raw_data.csv', mode='w', index=False)
```
