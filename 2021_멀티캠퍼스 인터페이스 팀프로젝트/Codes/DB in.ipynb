```python
gu_data = ['종로구','중구','용산구','성동구','광진구','동대문구','중랑구','성북구','강북구','도봉구','노원구','은평구','서대문구','마포구','양천구','강서구','구로구','금천구','영등포구','동작구','관악구','서초구','강남구','송파구','강동구']
gu_locx = [60, 60, 60, 61, 62, 61, 62, 61, 61, 61, 61, 59, 59, 59, 58, 58, 58, 59, 58, 59, 59, 61, 61, 62, 62]
gu_locy = [127, 127, 126, 127, 126, 127, 128, 127, 128, 129, 129, 127, 127, 127, 126, 126, 125, 124, 126, 125, 125, 125, 126, 126, 126]
gu_dict = {'gu_name': gu_data, 'gu_locx': gu_locx, 'gu_locy': gu_locy}

gu_data = pd.DataFrame(gu_dict)

data_set=[]
for a, b, c in gu_data.to_records(index=False):
    data_set.append((a, int(b), int(c)))
```

```python
# 구 데이터 insert
sql_g = """
    INSERT into tour_gu (gu_name, gu_locx, gu_locy)
    values (%s, %s, %s)
    """

curs.executemany(sql_g, tuple(data_set))
conn.commit()
```
